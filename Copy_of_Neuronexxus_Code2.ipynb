```python
!pip install gradio transformers torch librosa scikit-learn

import gradio as gr
import torch
import librosa
import numpy as np
from transformers import Wav2Vec2Processor, Wav2Vec2Model
from sklearn.ensemble import RandomForestClassifier
import joblib
# ----------------------------------
# Load Pretrained Model & Processor
# ----------------------------------

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base")
model.eval()

# ----------------------------------
# Load Trained Classifier or Simulate
# ----------------------------------

# Dummy classifier (for simulation purposes)
# Replace with: clf = joblib.load("depression_classifier.pkl") if you have one
clf = RandomForestClassifier()
X_dummy = np.random.rand(100, 768)
y_dummy = np.random.randint(0, 2, size=(100,))
clf.fit(X_dummy, y_dummy)

# ----------------------------------
# Prediction Function (60s Limit)
# ----------------------------------

def predict_depression(audio_file):
    try:
        # Load audio and check duration
        y, sr = librosa.load(audio_file, sr=16000)
        duration = librosa.get_duration(y=y, sr=sr)

        # Enforce 60-second limit
        if duration > 60:
            return f"Audio too long ({duration:.2f} sec). Please upload a clip shorter than 60 seconds."

        # Extract features using Wav2Vec2
        inputs = processor(y, sampling_rate=16000, return_tensors="pt", padding=True)
        with torch.no_grad():
            outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1).numpy()

        # Predict depression
        prediction = clf.predict(embedding)[0]
        prob = clf.predict_proba(embedding)[0][prediction]

        status = "Depressed" if prediction == 1 else "Not Depressed"
        return f"Prediction: {status}\nConfidence: {prob * 100:.2f}%"

    except Exception as e:
        return f"Error processing audio: {str(e)}"

# ----------------------------------
# Gradio Interface
# ----------------------------------

interface = gr.Interface(
    fn=predict_depression,
    inputs=gr.Audio(sources=["microphone", "upload"], type="filepath", label="Speak or Upload Audio"),
    outputs=gr.Textbox(label="Depression Detection Result"),
    title="Neuronexxus - Depression Detection",
    description="Speak or upload a short voice clip (max 60 seconds). The system simulates depression detection using Wav2Vec2 and a classifier.",
)

# ----------------------------------
# Launch Interface
# ----------------------------------

if __name__ == "__main__":
    interface.launch()
```

```python
import librosa
import soundfile as sf
import numpy as np
import os
from scipy.signal import butter, lfilter
# If you installed noisereduce, uncomment the line below:
# import noisereduce as nr

# --- Helper Functions (copy-paste from previous response) ---

def load_audio(file_path, target_sr=16000):
    """
    Loads an audio file and resamples it to the target_sr.
    """
    try:
        audio, sr = librosa.load(file_path, sr=target_sr) # Resamples to target_sr automatically
        return audio, sr
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None, None

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def apply_bandpass_filter(audio, sr, lowcut=300, highcut=3400): # Common speech frequencies
    b, a = butter_bandpass(lowcut, highcut, sr)
    filtered_audio = lfilter(b, a, audio)
    return filtered_audio

def remove_background_noise(audio, sr):
    """
    Applies a basic band-pass filter for noise reduction.
    For better results, consider using libraries like 'noisereduce' or
    more advanced signal processing techniques.
    """
    print("Applying basic noise reduction (band-pass filter). For better results, consider more advanced techniques.")
    processed_audio = apply_bandpass_filter(audio, sr)

    # Optional: If you want to try noisereduce (uncomment if installed)
    # try:
    #     print("Attempting noise reduction with 'noisereduce' library...")
    #     processed_audio = nr.reduce_noise(y=audio, sr=sr, stationary=True)
    # except Exception as e:
    #     print(f"Error with noisereduce: {e}. Falling back to band-pass filter.")
    #     processed_audio = apply_bandpass_filter(audio, sr)

    return processed_audio

def segment_audio(audio, sr, segment_length_sec=30, overlap_sec=0):
    """
    Segments audio into fixed-length chunks.
    segment_length_sec: desired length of each segment in seconds (e.g., 30 or 60)
    overlap_sec: overlap between segments in seconds (can be 0 for no overlap)
    """
    samples_per_segment = int(segment_length_sec * sr)
    overlap_samples = int(overlap_sec * sr)

    segments = []
    current_start = 0
    while current_start + samples_per_segment <= len(audio):
        segment = audio[current_start:current_start + samples_per_segment]
        segments.append(segment)
        current_start += samples_per_segment - overlap_samples

    # Handle the last segment if it's shorter than the desired length but still significant
    if len(audio) - current_start > samples_per_segment / 2: # Or some other threshold like 5 seconds of audio
        segment = audio[current_start:len(audio)]
        segments.append(segment)

    return segments

def preprocess_single_audio_file(file_path, target_sr=16000, segment_length_sec=30):
    """
    Orchestrates the preprocessing steps for a single audio file.
    """
    print(f"Processing: {file_path}")

    # 1. Load and normalize sampling rate
    audio_data, sr = load_audio(file_path, target_sr=target_sr)
    if audio_data is None:
        return [], None

    # 2. Remove background noise
    denoised_audio = remove_background_noise(audio_data, sr)

    # 3. Segment audio
    segmented_audios = segment_audio(denoised_audio, sr, segment_length_sec=segment_length_sec)

    print(f"Finished processing {os.path.basename(file_path)}. Created {len(segmented_audios)} segments.")
    return segmented_audios, sr
```

```python
import parselmouth
import numpy as np
import pandas as pd
import librosa

def extract_features(audio_file_path):
    """
    This function extracts a set of acoustic features from a given audio file.
    Features: Mean Pitch (F0), Jitter, Shimmer, and MFCCs.
    """
    try:
        # Load audio using both librosa and parselmouth
        y, sr = librosa.load(audio_file_path, sr=16000)
        sound = parselmouth.Sound(audio_file_path)

        # --- Pitch Extraction (using parselmouth) ---
        pitch = sound.to_pitch()

        # --- Jitter and Shimmer Extraction (using parselmouth) ---
        # THIS IS THE CORRECTED LINE: Use .get_floor() and .get_ceiling()
        point_process = parselmouth.praat.call(sound, "To PointProcess (periodic, cc)...", pitch.get_floor(), pitch.get_ceiling())

        jitter = parselmouth.praat.call(point_process, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
        shimmer = parselmouth.praat.call([sound, point_process], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)

        # Get mean pitch after creating the point_process
        pitch_values = pitch.selected_array['frequency']
        mean_pitch = np.mean(pitch_values[pitch_values > 0])


        # --- MFCC Extraction (using librosa) ---
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
        mean_mfccs = np.mean(mfccs, axis=1)

        # --- Combine all features ---
        feature_dict = {
            'mean_pitch_F0': mean_pitch,
            'jitter_local': jitter,
            'shimmer_local': shimmer,
        }
        for i, mfcc_val in enumerate(mean_mfccs):
            feature_dict[f'mfcc_{i+1}_mean'] = mfcc_val

        return feature_dict

    except Exception as e:
        print(f"Error processing {audio_file_path}: {e}")
        return None
```
